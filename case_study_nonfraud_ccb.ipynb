```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'Dispute Category': ['Unauthorized Transaction', 'Billing Error', 'Fraudulent Activity', 'Goods Not Received',
                         'Services Not Rendered', 'Duplicate Charge', 'Incorrect Amount', 'Credit not Processed',
                         'Card Not Present'],
    'Dispute Count': [4234, 2221, 2879, 9567, 550, 11816, 750, 95, 10],
    'Total Cost (USD)': [204875, 105679, 176446, 350439, 33879, 932706, 73725, 5804, 795]
}

df = pd.DataFrame(data)

# Plotting Dispute Count
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.bar(df['Dispute Category'], df['Dispute Count'], color='skyblue')
plt.xlabel('Dispute Category')
plt.ylabel('Dispute Count')
plt.title('Credit Card Disputes - Dispute Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Plotting Total Cost
plt.subplot(1, 2, 2)
plt.bar(df['Dispute Category'], df['Total Cost (USD)'], color='salmon')
plt.xlabel('Dispute Category')
plt.ylabel('Total Cost (USD)')
plt.title('Credit Card Disputes - Total Cost')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Show the plots
plt.show()
```
